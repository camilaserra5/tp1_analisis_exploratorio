```python
# librerías a utilizar
import numpy as np 
import pandas as pd
import datetime as datetime

# plots
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# cargamos los datos
events = pd.read_csv('events.csv', low_memory = False)
events['timestamp'] = pd.to_datetime(events['timestamp'], errors = 'coerce', format= '%Y-%m-%d')
events.head()
```

```python
# vemos que el país es sólo para el tipo de evento 'visited site'
events.groupby(['event', 'country']).agg({'count'}).head(10)
```

```python
# vemos los países desde los que más se visita la página (¿qué hacer con unknown?)
paises = events.groupby('country').agg({'event':'count'}).sort_values(by='event',ascending = False)[:10]
paises.reset_index(inplace=True)
paises.head()
```

```python
g = sns.barplot(y= paises.country, x=paises.event,  palette=sns.color_palette("husl", 8))
g.set_title("Visitas al sitio por pais", fontsize=15)
g.set_ylabel("Pais", fontsize=12)
g.set_xlabel("Cantidad", fontsize=12)
g.set_xticklabels([0,10000,20000,30000,40000,50000,60000,70000,80000],rotation = 45);
```

#### Podemos ver que la mayor cantidad de visitas al sitio ocurre en Brasil, analizaremos mas a fondo los datos que tenemos sobre este pais

```python
brazil_events = events[events['country'] == 'Brazil']
brazil_regions = brazil_events.groupby('region').agg({'event':'count'}).sort_values(by='event',ascending = False)
brazil_regions.reset_index(inplace=True)
brazil_regions = brazil_regions[brazil_regions['region'] != 'Unknown']
brazil_regions.head()
```

```python
g = sns.barplot(x= brazil_regions.event, y=brazil_regions.region[:15], palette=sns.color_palette("husl", 8))
g.set_title("Visitas por región (Brasil)", fontsize=15)
g.set_xlabel("Total de visitas", fontsize=12)
g.set_ylabel("", fontsize=1);
```

```python
# filtro por persona y región. cantidad de eventos únicos por region
persons_by_region = brazil_events.groupby(['region', 'person']).agg({'event':'count'}).sort_values(by='event',ascending = False).reset_index()
persons_by_region = persons_by_region.groupby('region').agg({'event':'count'}).sort_values(by='event',ascending = False).reset_index()
persons_by_region = persons_by_region[persons_by_region['region'] != 'Unknown']
persons_by_region.head()
```

```python
# agrego datos de población
persons_by_region['poblacion'] = [44169350, 16497395, 20777672, 15150143, 
                               9297861, 8867448, 11112062, 11228091, 
                               2867869, 6734568, 6551322, 6861924, 
                               8101180, 3894899, 3419550, 3950359, 
                               3893763, 3950359, 3198185, 3327551,
                               2630098, 1502759, 2227294, 1755015, 
                               756500, 795145, 500826]
persons_by_region.head()
```

Hay mucha más gente que viene de distrito federal (¿por que?)

```python
persons_by_region['events_por_poblacion'] = persons_by_region['event']/persons_by_region['poblacion']*100
persons_by_region = persons_by_region.sort_values(by='events_por_poblacion', ascending=False)
g = sns.barplot(x=persons_by_region.events_por_poblacion, y=persons_by_region.region[:15], palette=sns.color_palette("husl", 8))
g.set_title("Visitas por región (Brasil)", fontsize=15)
g.set_xlabel("Total de personas sobre población total", fontsize=12)
g.set_ylabel("Personas sobre población", fontsize=1);
```
